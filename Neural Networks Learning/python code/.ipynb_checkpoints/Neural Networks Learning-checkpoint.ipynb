## 1.3 Feedforward and cost function


### Compute Cost

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def sigmoidGradient(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

```python
def nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, lamb):
    #Reshape nn_params back into the parameters Theta1 and Theta2, the weight matrices
    #For our 2 layer neural network
    Theta1 = nn_params[:hidden_layer_size * (input_layer_size + 1)].reshape(hidden_layer_size, input_layer_size + 1)
    Theta2 = nn_params[hidden_layer_size * (input_layer_size + 1):].reshape(num_labels, hidden_layer_size + 1)
    #setup some useful variables
    m = X.shape[0]
    J = 0
    
    #add ones into matrix X
    X = np.append(np.ones((m, 1)), X, axis=1)
    
    #compute z2
    z2 = np.dot(X, Theta1.transpose())
    #compute a2
    a2 = sigmoid(z2)
    #add ones into matrix a2
    a2 = np.append(np.ones((m, 1)), a2, axis=1)
    
    #compute z3
    z3 = np.dot(a2, Theta2.transpose())
    #compute a3
    a3 = sigmoid(z3)
    
    #Converting y into vector of 0's and 1's for multi-class classification
    y_vec = np.zeros((m, num_labels))
    for i in range(m):
        y_vec[i, y[i] - 1] = 1
    
    #compute cost with regularization
    J = (1 / m) * np.sum(-y_vec * np.log(a3) - (1 - y_vec) * np.log(1 - a3)) + (lamb / (2 * m)) * (np.sum(np.square(Theta1[:, 1:])) + (np.sum(np.square(Theta2[:, 1:]))))
    
    
    #-------------------------------Backprobagation---------------------------
    #compute DELTA
    DELTA3 = a3 - y_vec # 5000 x 10
    DELTA2 = np.dot(DELTA3, Theta2) * np.append(np.ones((m, 1)), sigmoidGradient(z2), axis=1) # 5000 x 26
    DELTA2 = DELTA2[:, 1:] # 5000 X 25 (removing bias units)
    
    Theta1_grad = (1 / m) * np.dot(DELTA2.transpose(), X) # 25 x 401
    Theta2_grad = (1 / m) * np.dot(DELTA3.transpose(), a2)# 10 x 26
    
    #Regularization terms
    Theta1_grad_reg_term = (lamb / m) * np.append(np.zeros((hidden_layer_size, 1)), Theta1[:, 1:], axis=1) # 25 x 401
    Theta2_grad_reg_term = (lamb / m) * np.append(np.zeros((num_labels, 1)), Theta2[:, 1:], axis=1) # 10 x 26
    
    #adding Regularization terms in Theta1_grad and Theta2_grad
    Theta1_grad = Theta1_grad + Theta1_grad_reg_term
    Theta2_grad = Theta2_grad + Theta2_grad_reg_term
    
    #unroll parameters
    nn_params_grad = np.concatenate((Theta1_grad.flatten(), Theta2_grad.flatten()))
    #-------------------------------Backprobagation----------------------------
    
    
    return J, nn_params_grad
```

## 2.1 Sigmoid gradient

```python
def sigmoidGradient(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

## 2.2 Backpropagation algorithm

```python
def backpropagation(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, lamb):
    #Reshape nn_params back into the parameters Theta1 and Theta2, the weight matrices
    #For our 2 layer neural network
    Theta1 = nn_params[:hidden_layer_size * (input_layer_size + 1)].reshape(hidden_layer_size, input_layer_size + 1) # 25 x 401
    Theta2 = nn_params[hidden_layer_size * (input_layer_size + 1):].reshape(num_labels, hidden_layer_size + 1) # 10 x 26
    
    m = X.shape[0]
    
    #Add ones column into matrix X
    X = np.append(np.ones((m, 1)), X, axis=1) # 5000 x 401
    
    #compute z2
    z2 = np.dot(X, Theta1.transpose())
    #compute a2
    a2 = sigmoid(z2)
    #add ones column into matrix a2
    a2 = np.append(np.ones((m, 1)), a2, axis=1) # 5000 x 26
    
    #compute z3
    z3 = np.dot(a2, Theta2.transpose())
    #compute a3
    a3 = sigmoid(z3) # 5000 x 10
    
    #Converting y into vector of 0's and 1's for multi-class classification
    y_vec = np.zeros((m, num_labels)) # 5000 x 10
    for i in range(m):
        y_vec[i, y[i] - 1] = 1
    
    #compute DELTA
    DELTA3 = a3 - y_vec # 5000 x 10
    DELTA2 = np.dot(DELTA3, Theta2) * np.append(np.ones((m, 1)), sigmoidGradient(z2), axis=1) # 5000 x 26
    DELTA2 = DELTA2[:, 1:] # 5000 X 25 (removing bias units)
    
    Theta1_grad = (1 / m) * np.dot(DELTA2.transpose(), X) # 25 x 401
    Theta2_grad = (1 / m) * np.dot(DELTA3.transpose(), a2)# 10 x 26
    
    #Regularization terms
    Theta1_grad_reg_term = (lamb / m) * np.append(np.zeros((hidden_layer_size, 1)), Theta1[:, 1:], axis=1) # 25 x 401
    Theta2_grad_reg_term = (lamb / m) * np.append(np.zeros((num_labels, 1)), Theta2[:, 1:], axis=1) # 10 x 26
    
    #adding Regularization terms in Theta1_grad and Theta2_grad
    Theta1_grad = Theta1_grad + Theta1_grad_reg_term
    Theta2_grad = Theta2_grad + Theta2_grad_reg_term
    
    #unroll parameters
    nn_params_grad = np.concatenate((Theta1_grad.flatten(), Theta2_grad.flatten()))
    
    return nn_params_grad
```

## 2.3 Random initialization

```python
def randInitializeWeights(L_in, L_out):
    W = np.zeros((L_out, L_in + 1))
    
    #Randomly initialize the weights to small values
    epsilon_init = 0.12
    W = np.random.uniform(0, 1, size=(L_out, L_in + 1)) * epsilon_init * 2 - epsilon_init
    
    return W    
```

## 2.4 Gradient checking

```python
def debugInitializeWeights(fan_in, fan_out):
    W = np.sin(np.arange(1, 1 + (fan_out * (fan_in + 1))))
    return W.reshape(fan_out, fan_in + 1)
```

```python
def computeNumericalGradient(J, theta):
    e = 1e-4
    numgrad = np.zeros((theta.shape))
    perturb = np.diag(e * np.ones((theta.shape)))
    
    for i in range(len(theta)):
        loss1 = np.array(J(theta - perturb[:, i])[0])
        loss2 = np.array(J(theta + perturb[:, i])[0])
        #compute numerical gradient
        numgrad[i] = (loss2 - loss1) / (2*e)
        
    return numgrad
```

```python
def checkNNGradients(lamb):
    input_layer_size = 3
    hidden_layer_size = 5
    num_labels = 3
    m = 5
    
    #we generate some 'random' test data
    Theta1 = debugInitializeWeights(input_layer_size, hidden_layer_size)
    Theta2 = debugInitializeWeights(hidden_layer_size, num_labels)
    
    #Reusing debugInitializeWeights to generate X
    X = debugInitializeWeights(input_layer_size - 1, m)
    y = np.random.randint(0, num_labels, size=(m))
    
    #unroll parameters
    nn_params = np.concatenate((Theta1.flatten(), Theta2.flatten()))
    
    #short hand for cost function
    costFunc = lambda p: nnCostFunction(p, input_layer_size, hidden_layer_size, num_labels, X, y, lamb)
    
    cost, grad = costFunc(nn_params)
    numgrad = computeNumericalGradient(costFunc, nn_params)
    
    #Visually examine the two gradient computations. The two columns you get should be very similar.
    print(np.stack([numgrad, grad], axis=1))
    print('The above two columns you get should be very similar')
    print('Left-Your Numerical Gradient, Right-Analytical Gradient')
    
    # Evaluate the norm of the difference between two the solutions. If you have a correct
    # implementation, and assuming you used EPSILON = 0.0001 in computeNumericalGradient, then diff
    # should be less than 1e-9.
    diff = np.linalg.norm(numgrad - grad) / np.linalg.norm(numgrad + grad)
    
    print('If your backpropagation implementation is correct, then \n'
          'the relative difference will be small (less than 1e-9). \n'
          'Relative Difference: ', diff)
```

```python
#check gradients by running checkNNGradients
lamb = 3
checkNNGradients(lamb)
```

# 3. Neural Networks Training

```python
#predict function
def predict(Theta1, Theta2, X):
    m = X.shape[0]
    X = np.append(np.ones((m, 1)), X, axis=1)
    
    #compute z2
    z2 = np.dot(X, Theta1.transpose())
    #compute a2
    a2 = sigmoid(z2)
    #add ones column into matrix a2
    a2 = np.append(np.ones((m, 1)), a2, axis=1)
    
    #compute z3
    z3 = np.dot(a2, Theta2.transpose())
    #compute a3
    a3 = sigmoid(z3)
    
    return np.argmax(a3, axis=1) + 1
```

## 3.1 Learning parameters using Gradient Descent

```python
def nnGradientDescent(X, y, initial_nn_params, alpha, num_iters, lamb, input_layer_size, hidden_layer_size, num_labels):
    Theta1 = initial_nn_params[:hidden_layer_size * (input_layer_size + 1)].reshape(hidden_layer_size, input_layer_size + 1)
    Theta2 = initial_nn_params[hidden_layer_size * (input_layer_size + 1):].reshape(num_labels, hidden_layer_size + 1)
    
    m = len(y)
    J_history = []
    
    for i in range(num_iters):
        nn_params = np.concatenate((Theta1.flatten(), Theta2.flatten()))
        #calculate cost and theta gradient
        cost, grad = nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, lamb)
        #calculate theta1 and theta2
        Theta1 = Theta1 - alpha * grad[:hidden_layer_size * (input_layer_size + 1)].reshape(hidden_layer_size, input_layer_size + 1)
        Theta2 = Theta2 - alpha * grad[hidden_layer_size * (input_layer_size + 1):].reshape(num_labels, hidden_layer_size + 1)
        #adding cost to J_history list
        J_history.append(cost)
        
    nn_grad_params = np.concatenate((Theta1.flatten(), Theta2.flatten()))
    return nn_grad_params, J_history
```

```python
lamb = 1
alpha = 0.8
num_iters = 1000
initial_nn_params = np.concatenate((initial_Theta1.flatten(), initial_Theta2.flatten()))
```

```python
nnTheta, nnJ_history = nnGradientDescent(X, y, initial_nn_params, alpha, num_iters, lamb, input_layer_size, hidden_layer_size, num_labels)
```

```python
print('error: ', nnJ_history[len(nnJ_history) - 1])
```

## 3.2 Learning parameters using scipy.optimize.minimize

```python
import scipy.optimize as op
```

```python
options={'maxiter': 800}
lamb = 1

# Create "short hand" for the cost function to be minimized
costFunction = lambda p: nnCostFunction(p, input_layer_size, hidden_layer_size, num_labels, X, y, lamb)

# Now, costFunction is a function that takes in only one argument
# (the neural network parameters)
optimization = op.minimize(costFunction, initial_nn_params, jac=True, method='CG', options=options)
```

```python
# get the solution of the optimization
op_params = optimization.x
```

```python
# Obtain Theta1 and Theta2 back from nn_params
op_Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)], (hidden_layer_size, (input_layer_size + 1)))

op_Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):], (num_labels, (hidden_layer_size + 1)))
```

```python
op_predictions = predict(op_Theta1, op_Theta2, X)
```

```python
print('error: ', optimization.fun)
```

```python
print('Training set Accuracy: ', np.mean(op_predictions.reshape(y.shape) == y)*100, '%')
```
